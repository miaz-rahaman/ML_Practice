```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
print('Dataset Priview')
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='scatter')  # for kind we have the following 'scatter', 'hist', 'hex', 'kde', 'reg', 'resid'
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.distplot(df['tip'])
```

# `distplot` is a deprecated function and will be removed in seaborn v0.14.0.

Please adapt your code to use either `displot` (a figure-level function with

```python
sns.countplot(x='smoker',data=df)
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df)
```

# Violin Plot
this helps us see both the distribution of data in terms of kernal density and also box plot

```python
sns.violinplot(x='total_bill',y='day',hue='sex',data=df,palette='rainbow')
```
